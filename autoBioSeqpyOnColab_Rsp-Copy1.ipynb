```python
dataEncodingType = 'dict'
#dataEncodingType = 'onehot'

useKMer = True
#useKMer = False
KMerNum = 3 #If useKMer is False, the KMerNum is inactive, and thus it doesn't matter how much the value is.

spcLen = 100
#spcLen = 50
```

```python
featureGenerator = dataProcess.DNAFeatureGenerator(dataEncodingType, useKMer=useKMer, KMerNum=KMerNum) #alternative for DNA data
```

```python
#the paths
#dataTrainFilePaths = ['C:\\hmm\\courses\\Research Work\\Data\\Rhodobactor S\\positive R data\\pos R data.fna']
#dataTrainFilePaths = ['C:\\hmm\\courses\\Research Work\\Data\\Rhodobactor S\\negative R data\\neg R data1.fna']
dataTrainFilePaths = ['C:\\hmm\\courses\\Research Work\\Data\\Rhodobactor S\\positive R data\\pos R data.fna', 'C:\\hmm\\courses\\Research Work\\Data\\Rhodobactor S\\negative R data\\neg R data1.fna']

#the related labels
dataTrainLabel = [1, 0]
#a list for recording the DataLoader
trainDataLoaders = []
for i,dataPath in enumerate(dataTrainFilePaths):
    #init
    dataLoader = dataProcess.DataLoader(label = dataTrainLabel[i], featureGenerator=featureGenerator)
    #file read
    dataLoader.readFile(dataPath, spcLen = spcLen)
    trainDataLoaders.append(dataLoader)

#init
trainDataSetCreator = dataProcess.DataSetCreator(trainDataLoaders)

```

```python
dataSplitScale = 0.8
trainDataMat, testDataMat, trainLabelArr, testLabelArr = trainDataSetCreator.getTrainTestSet(dataSplitScale, toShuffle=True)

print('Training:')
print('Matrix with shape %d x %d:' %(trainDataMat.shape[0],trainDataMat.shape[1]))
print(trainDataMat)
print('\n')
print('The labels with length %d:' %(len(trainLabelArr)))
print(trainLabelArr)
print('\n########################################################################################')
print('########################################################################################\n')
print('Testing:')
print('Matrix with shape %d x %d:' %(testDataMat.shape[0],testDataMat.shape[1]))
print(testDataMat)
print('\n')
print('The labels with length %d:' %(len(testLabelArr)))
print(testLabelArr)
```

```python
#===================
#DNA_dict_model.py
#===================

def DNA_dict_model():

  from keras.models import Sequential
  from keras.layers import Dense, Dropout, Activation, Flatten
  from keras.layers import Embedding
  from keras.layers import Conv1D, AveragePooling1D
  from keras import optimizers

  filters = 80
  kernel_size = 4 
  pool_size = 2
  hidden_dims1 = 80
  hidden_dims2 = 40
  hidden_dims3 = 40

  if useKMer:
    if KMerNum == 1:
      max_features = 5
      embedding_size = 16
    elif KMerNum == 2:
      max_features = 25
      embedding_size = 64
    elif KMerNum == 3:
      max_features = 125
      embedding_size = 256
  else:
    max_features = 5
    embedding_size = 16

  maxlen = trainDataMat.shape[1]

  ##the parameter which need to modified
  #if useKMer:
  #    maxlen = spcLen - KMerNum + 1
  ##    max_features = 4 ** KMerNum
  #    max_features = trainDataMat.shape[1]
  #else:
  #    maxlen = spcLen
  #    max_features = trainDataMat.shape[1]

  print('Building model...')
  model = Sequential()
  # here you need to specify the input_length, we set the value of maxlen as an example
  model.add(Embedding(max_features, embedding_size, input_length=maxlen))
  model.add(Conv1D(filters,kernel_size = kernel_size, padding ='valid', activation = 'relu', strides = 1))
  model.add(AveragePooling1D(pool_size = pool_size))
  model.add(Flatten())
  model.add(Dropout(0.3))
  model.add(Dense(hidden_dims1))
  model.add(Dropout(0.3))
  model.add(Dense(hidden_dims2))
  model.add(Dropout(0.3))
  model.add(Dense(hidden_dims3))
  model.add(Dropout(0.3))
  model.add(Activation('relu'))
  model.add(Dense(1))
  model.add(Activation('sigmoid'))

  model.compile(loss = 'binary_crossentropy',optimizer = optimizers.Adam(),metrics = ['acc'])

  model.summary()

  return model
```

```python
#===================
#DNA_onehot_model.py
#===================

def DNA_onehot_model():

  from keras.models import Sequential
  from keras.layers import Dense, Dropout, Activation, Flatten, Reshape
  from keras.layers import Conv1D, AveragePooling1D
  from keras import optimizers

  filters = 80
  kernel_size = 5 
  pool_size = 2
  hidden_dims1 = 80
  hidden_dims2 = 40
  hidden_dims3 = 40
 
  #the parameter which need to modified
  if useKMer:
    maxlen = spcLen - KMerNum + 1
  #  max_features = 5 ** KMerNum
    max_features = trainDataMat.shape[1]
    col = 4 ** KMerNum
    row = int(max_features / col)
  else:
    maxlen = spcLen
#    max_features = 5
    max_features = trainDataMat.shape[1]
    col = 4
    row = int(max_features / col)

  print('Building model...')
  model = Sequential()
  model.add(Reshape((row,col),input_shape=(max_features,)))
  model.add(Conv1D(filters,kernel_size = kernel_size,padding ='valid',activation = 'relu',strides = 1))
  model.add(AveragePooling1D(pool_size = pool_size))
  model.add(Flatten())
  model.add(Dropout(0.3))
  model.add(Dense(hidden_dims1))
  model.add(Dropout(0.3))
  model.add(Dense(hidden_dims2))
  model.add(Dropout(0.3))
  model.add(Dense(hidden_dims3))
  model.add(Dropout(0.3))
  model.add(Activation('relu'))
  model.add(Dense(1))
  model.add(Activation('sigmoid'))

  model.compile(loss = 'binary_crossentropy',optimizer = optimizers.Adam(),metrics = ['acc'])

  model.summary()

  return model
```

```python

if dataEncodingType == 'dict':
  model = DNA_dict_model()
else:
  model = DNA_onehot_model()
```

```python
import analysisPlot

batch_size = 40
epochs = 20

history = analysisPlot.LossHistory()
model.fit(trainDataMat, trainLabelArr, batch_size = batch_size, epochs = epochs, validation_split = 0.5, callbacks = [history])
```

```python
predicted_Probability = model.predict(testDataMat)
#HC commented out...
#prediction = model.predict_classes(testDataMat)

#prediction = (model.predict(testDataMat) > 0.5).astype("int32")
prediction = (predicted_Probability > 0.5).astype("int32")
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, recall_score, precision_score, confusion_matrix, matthews_corrcoef, auc, roc_curve
```

```python
history.loss_plot('epoch',showFig=True,savePath=None)
```

```python
cm=confusion_matrix(testLabelArr,prediction)
print(cm)
print("ACC: %f "%accuracy_score(testLabelArr,prediction))
print("F1: %f "%f1_score(testLabelArr,prediction))
print("Recall: %f "%recall_score(testLabelArr,prediction))
print("Pre: %f "%precision_score(testLabelArr,prediction))
print("MCC: %f "%matthews_corrcoef(testLabelArr,prediction))
fpr,tpr,threshold = roc_curve(testLabelArr, predicted_Probability)
auc_roc = auc(fpr,tpr)
print("AUC: %f "%auc_roc)
```
